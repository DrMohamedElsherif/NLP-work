# Projektbeschreibung: Relation Extraction

Das Projekt besteht darin, einen Klassifikator zu entwickeln, der entscheiden kann, welcher Relationstyp zwischen zwei Nomen in einem Satz vorliegt. 

## Daten

Für die Entwicklung des Klassifikators nutzen wir die Daten des SemEval 2010 Task 8, ein Korpus mit 10717 annotierten Sätzen: 

- Informationen zum Korpus: https://aclanthology.org/S10-1006/
- Corpus als zip-Datei: https://moodle.zdv.uni-tuebingen.de/pluginfile.php/22310/mod_folder/content/0/SemEval2010_task8.zip

Bitte entpacken Sie das Corpus im Ordner ihres Notebooks.

Das Corpus liegt im CoNLL-Format vor. Jede Zeile enthält eine ID, das Label des Relationstyps und schließlich den Satz, wobei die Felder jeweils durch einen Tabulator getrennt sind:

    1	Component-Whole(e2,e1)	The system as described above has its greatest application in an arrayed <e1>configuration</e1> of antenna <e2>elements</e2>.
    
Die beiden Nomen sind durch `<e1>...</e1>` und `<e2>...</e2>` markiert. Die **Menge der Relationstypen** ist vorgegeben und hier beschrieben: 

- `Cause-Effect(X, Y)` is true for a sentence S that mentions entities X and Y if and only if the situation described in S entails that X is the cause of Y, or that X causes/makes/produces/emits/... Y.
- `Instrument-Agency(X, Y)` is true of a sentence S that mentions entities X and Y if and only if the situation described in S entails the fact that X is the instrument (tool) of Y or, equivalently, that Y uses X.
- `Product-Producer (X, Y)` is true for a sentence S that mentions entities X and Y if and only if the situation described in S entails the fact that X is a product of Y, or Y produces X.
- `Content-Container(X, Y)` is true for a sentence S that mentions entities X and Y if and only if the situation described in S entails that X is or was (usually temporarily) stored or carried inside Y.
- `Entity-Origin(X, Y)` is true for a sentence S that mentions the entities X and Y if and only if the situation described in S entails that Y is the origin of an entity X (rather than its location), and X is coming or derived from that origin.
- `Entity-Destination(X, Y)` is true for a sentence S that mentions the entities X and Y if and only if the situation described in S entails the fact that Y is the destination of X in the sense of X moving (in a physical or abstract sense) toward Y.
- `Component-Whole (X,Y)` is true for a sentence S that mentions entities X and Y if and only if the situation described in S entails that X is a component of Y, and X has a functional relation with Y. In other words, X has an operating or usable purpose within Y.
- `Member-Collection(X, Y)` is true for a sentence S that mentions entities X and Y if and only if the situation described in S entails the fact that X is a member of Y.
- `Message-Topic(X, Y)` is true for a sentence S that mentions the entities X and Y if and only if the situation described in S entails the fact that X is a communicative message containing information about Y.

```python
# Calculate confidence intervals
def confidence_interval(data, confidence=0.95):
    n = len(data)
    if n <= 1:                              # This will Handle cases with insufficient data
        return np.nan, np.nan
    mean = np.mean(data)
    std_err = sem(data, nan_policy="omit")  # This will Handle NaN values (if present)
    h = std_err * t.ppf((1 + confidence) / 2, n - 1)
    return mean, h

# Analyze data
def analyze_data(df, name):
    print(f"\nStatistical Analysis for {name} Data:")
    print("-------------------------------------------")
    print(f"Number of Instances: {len(df)}")
    
    # Average Sentence Length
    df['Tokenized_Sentence'] = df['Sentence'].apply(word_tokenize)
    df['Sentence_Length'] = df['Tokenized_Sentence'].apply(len)
    print(f"Average Sentence Length: {df['Sentence_Length'].mean():.2f} Words")

    # Standard Deviation
    print(f"Standard Deviation of Sentence Length: {df['Sentence_Length'].std():.2f}")

    # Outlier Detection
    Q1 = df["Sentence_Length"].quantile(0.25)
    Q3 = df["Sentence_Length"].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df["Sentence_Length"] < (Q1 - 1.5 * IQR)) | (df["Sentence_Length"] > (Q3 + 1.5 * IQR))]
    print(f"Number of Outliers: {len(outliers)}")

# Analyze train and test data
analyze_data(train, "Training")
analyze_data(test, "Test")

# Average Sentence Length Per Class
train["Sentence Length"] = train["Sentence"].apply(lambda x: len(word_tokenize(x)))
test["Sentence Length"] = test["Sentence"].apply(lambda x: len(word_tokenize(x)))

train_avg_length = train.groupby("Relation")["Sentence Length"].mean()
test_avg_length = test.groupby("Relation")["Sentence Length"].mean()

train_median_length = train.groupby("Relation")["Sentence Length"].median()
test_median_length = test.groupby("Relation")["Sentence Length"].median()

train_ci = train.groupby("Relation")["Sentence Length"].apply(confidence_interval)
test_ci = test.groupby("Relation")["Sentence Length"].apply(confidence_interval)

# Plot Sentence Length Distributions (Train & Test)
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Train Set
sns.violinplot(
    x=train["Relation"],
    y=train["Sentence Length"],
    hue=train["Relation"],  
    palette="coolwarm",
    ax=axes[0],
    inner="box",
    legend=False  
)
axes[0].scatter(x=range(len(train_avg_length)), y=train_avg_length, color="black", marker="D", s=50, label="Mean Length", zorder=2)
axes[0].scatter(x=range(len(train_median_length)), y=train_median_length, color="red", marker="o", s=50, label="Median Length", zorder=2)
axes[0].set_title("Sentence Length Distribution in Training Set", fontsize=14, pad=20)
axes[0].xaxis.set_major_locator(FixedLocator(range(len(train_avg_length))))
axes[0].set_xticklabels(train_avg_length.index, rotation=90)
axes[0].set_xlabel("Relation")
axes[0].set_ylabel("Sentence Length (Tokens)")
axes[0].legend()

# Test Set
sns.violinplot(
    x=test["Relation"],
    y=test["Sentence Length"],
    hue=test["Relation"],  
    palette="coolwarm",
    ax=axes[1],
    inner="box",
    legend=False  
)
axes[1].scatter(x=range(len(test_avg_length)), y=test_avg_length, color="black", marker="D", s=50, label="Mean Length", zorder=2)
axes[1].scatter(x=range(len(test_median_length)), y=test_median_length, color="red", marker="o", s=50, label="Median Length", zorder=2)
axes[1].set_title("Sentence Length Distribution in Test Set", fontsize=14, pad=20)
axes[1].xaxis.set_major_locator(FixedLocator(range(len(test_avg_length))))
axes[1].set_xticklabels(test_avg_length.index, rotation=90)
axes[1].set_xlabel("Relation")
axes[1].set_ylabel("Sentence Length (Tokens)")
axes[1].legend()

plt.tight_layout()
plt.show()
```

#### 3.1. **Default-Baseline Model**

```python
def baseline_model(train, test):
    # Identify the majority class
    majority_class = train['Relation'].mode()[0]
    print(f'Majority Class: {majority_class}')
    
    # Create predictions based on the majority class
    predictions = [majority_class] * len(test)
    
    # Calculate accuracy
    dummy_accuracy = accuracy_score(test['Relation'], predictions)
    print("-------------------------------")
    print(f"Default-Baseline Model Accuracy: {dummy_accuracy * 100:.2f}%")
    print("-------------------------------")
    
    # Plot confusion matrix
    cm = confusion_matrix(test['Relation'], predictions, labels=train['Relation'].unique())
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=train['Relation'].unique(), yticklabels=train['Relation'].unique())
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.title('Confusion Matrix - Default Baseline')
    plt.show()


baseline_model(train, test)
```

Now, to build a Unigramm-Baselime Model, I encode the target labels using LabelEncoder to convert them into numerical format. Next, I transform the sentences into Bag-of-Words (BoW) features using CountVectorizer, which creates a sparse matrix of word frequencies. This will be beneficial later when training the SVM model. 

<div class="alert alert-block alert-success"> 
    
#### 3.2. **Unigramm-Baseline Model**
</div>

```python
def Unigram_BaselineNB(train, test):
    # Preprocess text 
    def preprocess_text(sentence):
        sentence = re.sub(r"</?e\d>", "", sentence)  # This will remove entity markers 
        return sentence.lower()                      # and have case-insensitivity 

    # Apply preprocessing to train and test sets
    train["Processed"] = train["Sentence"].apply(preprocess_text)
    test["Processed"] = test["Sentence"].apply(preprocess_text)

    # Convert text into unigram features (Bag-of-Words)
    vectorizer = CountVectorizer(ngram_range=(1, 1))       # Unigram model 
    X_train = vectorizer.fit_transform(train["Processed"])
    X_test = vectorizer.transform(test["Processed"])

    # Encode labels using LabelEncoder
    label_encoder = LabelEncoder()
    y_train = label_encoder.fit_transform(train["Relation"])
    y_test = label_encoder.transform(test["Relation"])

    # Train the Naive Bayes classifier
    nb_classifier = MultinomialNB()  
    nb_classifier.fit(X_train, y_train)

    # Make predictions
    y_pred = nb_classifier.predict(X_test)

    # Evaluate the model
    nb_accuracy = accuracy_score(y_test, y_pred)
    print("------------------------------------")
    print(f"Unigram Naive Bayes Model Accuracy: {nb_accuracy * 100:.2f}%")
    print("------------------------------------")

    # Confusion Matrix
    conf_matrix = confusion_matrix(y_test, y_pred)

    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Oranges",
                xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title("Confusion Matrix - Unigram Naive Bayes")
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.show()

Unigram_BaselineNB(train, test)
```

<div class="alert alert-block" style="background-color: #ffcc99; border-color: #ff9933; color: #000000;">

Next, I extract the following features from each sentence:

- The first verb in the sentence
- The target nouns <e1> and <e2>
- A preposition between E1 and E2, if one exists

To achieve this, I perform tokenization and POS tagging as below.

<div class="alert alert-block alert-success"> 
    
#### 4.1. **Wortform eines Verbs im Satz**
#### 4.2. **Wortform der Targetnomen E1 & E2**
#### 4.3. **Wortform einer Präsposition zwischen den Targetnomen (E1 & E2)**
</div>

```python
# Function to extract features: Verb, E1, E2, Preposition
def extract_features(sentence):
    tokens = word_tokenize(sentence)
    pos_tags = pos_tag(tokens)

    # Extract first verb
    verb = next((word.lower() for word, tag in pos_tags if tag.startswith("VB")), "")

    # Extract <e1> and <e2> words
    e1_match = re.search(r"<e1>(.*?)</e1>", sentence)
    e2_match = re.search(r"<e2>(.*?)</e2>", sentence)
    e1_word = e1_match.group(1).lower() if e1_match else ""
    e2_word = e2_match.group(1).lower() if e2_match else ""

    # Extract preposition between E1 and E2
    between_prep = ""
    if e1_match and e2_match:
        e1_start, e1_end = e1_match.span()
        e2_start, e2_end = e2_match.span()
        if e1_end < e2_start:
            between_words = word_tokenize(sentence[e1_end:e2_start])
        else:
            between_words = word_tokenize(sentence[e2_end:e1_start])   # If E2 appears before E1

        # Find first preposition (IN tag)
        prepositions = [word.lower() for word, tag in pos_tag(between_words) if tag.startswith("IN")]
        if prepositions:
            between_prep = prepositions[0]

    return verb, e1_word, e2_word, between_prep

# Apply feature extraction
train[["Verb", "E1", "E2", "Preposition"]] = train["Sentence"].apply(lambda x: pd.Series(extract_features(x)))
test[["Verb", "E1", "E2", "Preposition"]] = test["Sentence"].apply(lambda x: pd.Series(extract_features(x)))
```

<div class="alert alert-block alert-success"> 
    
#### 4.4. **Word2Vec-Embedding**
</div>

```python
# Train Word2Vec Model on both Train and Test data to cover all words
sentences = [word_tokenize(sent.lower()) for sent in pd.concat([train["Sentence"], test["Sentence"]])]
w2v_model = Word2Vec(sentences, vector_size=100, window=5, min_count=2, workers=8, sg=1)  # min_count=2 to ignore rare words that don’t contribute 
                                                                                          # meaningful embeddings

# Function to get Word2Vec embeddings
def get_w2v_vector(word):
    return w2v_model.wv[word] if word in w2v_model.wv else np.zeros(100)

# Convert E1 and E2 into their embeddings
train["W2V_E1"] = train["E1"].apply(lambda x: get_w2v_vector(x))
train["W2V_E2"] = train["E2"].apply(lambda x: get_w2v_vector(x))
test["W2V_E1"] = test["E1"].apply(lambda x: get_w2v_vector(x))
test["W2V_E2"] = test["E2"].apply(lambda x: get_w2v_vector(x))

# Convert Text into Bag-of-Words Features
vectorizer = CountVectorizer()
X_train_text = vectorizer.fit_transform(train["Sentence"])
X_test_text = vectorizer.transform(test["Sentence"])

# Stack all features 
X_train = np.hstack((X_train_text.toarray(), np.vstack(train["W2V_E1"]), np.vstack(train["W2V_E2"])))
X_test = np.hstack((X_test_text.toarray(), np.vstack(test["W2V_E1"]), np.vstack(test["W2V_E2"])))

# Encode Labels
label_mapping = {label: idx for idx, label in enumerate(train["Relation"].unique())}
y_train = train["Relation"].map(label_mapping)
y_test = test["Relation"].map(label_mapping)
```

<div class="alert alert-block" style="background-color: #ffcc99; border-color: #ff9933; color: #000000;">

Now, I train a Support Vector Machine model using all the features I extracted so far. 

<div class="alert alert-block alert-success"> 
    
#### 4.5. **Support Vector Machines (SVM) classification**
</div>

```python
# Train SVM Model
SVM_classifier = LinearSVC(C=1.0, max_iter=5000, random_state=42, dual=False, tol=1e-4, class_weight= None)
print("Training SVM model...")
with tqdm(total=100) as pbar:
    SVM_classifier.fit(X_train, y_train)
    pbar.update(100)

y_pred = SVM_classifier.predict(X_test)
```

<div class="alert alert-block alert-success"> 
    
##### 4.5.1 **Evaluation of the Support Vector Machines (SVM) Model**
</div>

```python
# Evaluation
accuracy = accuracy_score(y_test, y_pred)
print("\n---------------------")
print(f"SVM Model Accuracy: {accuracy * 100:.2f}%")
print("---------------------")

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 5))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Purples", xticklabels=label_mapping.keys(), yticklabels=label_mapping.keys())
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - SVM Model")
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()

# Classification Report
print("\n------------------------")
print("Classification Report:")
print("------------------------")
print(classification_report(y_test, y_pred, target_names=label_mapping.keys(), zero_division=0))

# Calculate per-class accuracy
per_class_accuracy = {}
for label in label_mapping.keys():
    idx = label_mapping[label]
    mask = y_test == idx
    if sum(mask) > 0:   # This to avoid the warning message due to division by zero
        per_class_accuracy[label] = accuracy_score(y_test[mask], y_pred[mask])

print("\n--------------------")
print("Per-Class Accuracy:")
print("--------------------")
for label, acc in per_class_accuracy.items():
    print(f"{label}: {acc * 100:.2f}%")
```

```python
# Function to get WordNet synonyms
def get_synonyms(word):
    synonyms = set()
    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())
    return " ".join(synonyms)

# Add WordNet features to the dataset
train["Synonyms_E1"] = train["E1"].apply(get_synonyms)
train["Synonyms_E2"] = train["E2"].apply(get_synonyms)
test["Synonyms_E1"] = test["E1"].apply(get_synonyms)
test["Synonyms_E2"] = test["E2"].apply(get_synonyms)

# Vectorize WordNet features
syn_vectorizer = CountVectorizer()
X_train_syn = syn_vectorizer.fit_transform(train["Synonyms_E1"] + " " + train["Synonyms_E2"])
X_test_syn = syn_vectorizer.transform(test["Synonyms_E1"] + " " + test["Synonyms_E2"])

# Combine with all existing features
X_train_combined = np.hstack((X_train, X_train_syn.toarray()))
X_test_combined = np.hstack((X_test, X_test_syn.toarray()))

# Train the SVM model with the added WordNet features
print("Training SVM model with WordNet features...")
with tqdm(total=100) as pbar:
    SVM_classifier.fit(X_train_combined, y_train)
    pbar.update(100)

# Make Predictions on the test set
y_pred_combined = SVM_classifier.predict(X_test_combined)

# Evaluate the model
accuracy_combined = accuracy_score(y_test, y_pred_combined)
print(f"SVM Model Accuracy (with WordNet Features): {accuracy_combined * 100:.4f}%")

# Confusion Matrix
conf_matrix_combined = confusion_matrix(y_test, y_pred_combined)
plt.figure(figsize=(8, 5))
sns.heatmap(conf_matrix_combined, annot=True, fmt="d", cmap="Purples", xticklabels=label_mapping.keys(), yticklabels=label_mapping.keys())
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - SVM Model (with WordNet Features)")
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()

# Classification Report
print("\n------------------------")
print("Classification Report (with WordNet Features):")
print("------------------------")
print(classification_report(y_test, y_pred_combined, target_names=label_mapping.keys(), zero_division=0))

# Calculate per-class accuracy with WordNet features
per_class_accuracy_combined = {}
for label in label_mapping.keys():
    idx = label_mapping[label]
    mask = y_test == idx
    if sum(mask) > 0:  # Avoid division by zero
        per_class_accuracy_combined[label] = accuracy_score(y_test[mask], y_pred_combined[mask])

# Print per-class accuracy
print("\n--------------------")
print("Per-Class Accuracy (with WordNet Features):")
print("--------------------")
for label, acc in per_class_accuracy_combined.items():
    print(f"{label}: {acc * 100:.2f}%")
```

```python
# Define parameters 
param_grid = {
    'C': [0.001, 0.01, 0.1, 1.0, 10.0],  
    'max_iter': [1000, 2000, 5000],  
    'tol': [1e-4, 1e-5, 1e-3],   
    'class_weight': [None, 'balanced'],  
}


# Grid search with cross-validation
print("Performing Grid Search with Cross-Validation...")
grid_search = GridSearchCV(
    estimator=LinearSVC(dual=False, random_state=42),  
    param_grid=param_grid,
    scoring='accuracy',   
    cv=3,       # Here, I used 3-fold instead of 5-fold because some classses has very low number of instances. Using smaller CV                 
    n_jobs=-1,  # improved the performance alittle bit!
    verbose=0  
)

# Follow the progress of the CV 
with tqdm(total=len(param_grid['C']) * len(param_grid['max_iter']) * len(param_grid['class_weight'])) as pbar:
    grid_search.fit(X_train_combined, y_train)
    pbar.update(len(param_grid['C']) * len(param_grid['max_iter']) * len(param_grid['class_weight']))

# Print best parameters and accuracy
print("\nBest Parameters Found:")
print(grid_search.best_params_)
print(f"Best Cross-Validation Accuracy: {grid_search.best_score_ * 100:.4f}%")

# Train the model with the best performing parameters
best_svm = grid_search.best_estimator_
print("\nTraining the model with the best parameters...")
with tqdm(total=100) as pbar:
    best_svm.fit(X_train_combined, y_train)
    pbar.update(100)
```
